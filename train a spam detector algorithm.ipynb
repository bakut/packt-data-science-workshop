Train a spam detetctor algorithm

```
import pandas as pd
```

```
csv_url = ('https://raw.githubusercontent.com/PacktWorkshops/The-Data-Science-Workshop/master/Chapter01/Dataset/dataset_44_spambase.csv')
```

```
csv_df = pd.read_csv(csv_url)
```

```
csv_df
```

```
#extrat the class target variable using .pop() method from pandas and savw the result in variable called target:
target = csv_df.pop('class')
```

```
print(target)
```

```
#import the RandomForestClassifier class from sklearn.ensemble
from sklearn.ensemble import RandomForestClassifier
```

```
seed = 168
```

```
#instantiate randomforestclassifier with the random_state_seed parameter and save it into variable called rf_model
rf_model = RandomForestClassifier(random_state=seed)
```

```
#train the model with the .fit() method with df and target as parameters
rf_model.fit(csv_df, target)
```

```
#making predictions
preds= rf_model.predict(csv_df)
```

```
print(preds)
```

```
#imort the accuracy_score method from sklearn.metrics:
from sklearn.metrics import accuracy_score
```

```
#calculate accuracy_score() with targets and preds as parameters and save result in variables called acc_score:
acc_score = accuracy_score(target, preds)
```

```
print(acc_score)
```
